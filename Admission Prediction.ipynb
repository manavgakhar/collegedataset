```python
df.drop(['Serial No.'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.columns
```

```python
lis=['GRE Score', 'TOEFL Score', 'SOP', 'LOR ', 'CGPA']
for i in lis:
    scale=StandardScaler()
    df[i]= scale.fit_transform(np.array(df[i]).reshape(-1,1))
    
```

```python
df.head()
```

```python
a= pd.get_dummies(df['University Rating'])
```

```python
fin_df= pd.concat([df,a],axis=1)
```

```python
fin_df.drop(['University Rating'],axis=1,inplace=True)
```

```python
fin_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
fin_df.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(fin_df.drop(['Chance of Admit '],axis=1), fin_df['Chance of Admit '], test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred= model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```

```python
#This is a good r2 score
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,pred)
```

```python
sns.distplot(abs(y_test-pred),bins=30)
```

```python
# Mean absolute error is low and distplot shows most errors lie in that range as well. Even outliers get capped at 0.3.
```
